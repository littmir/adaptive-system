**Простые действия с массивами:**

```python
a = np.array([1, 2])
b = np.array([2, 3])
```

```python
#сложение
print(a + b, '\n')
print(np.add(a, b))
```

```python
#вычитание
print(a - b, '\n')
print(np.subtract(a, b))
```

```python
#деление
print(a / b, '\n')
print(np.divide(a, b))
```

```python
#поэлементное умножение
print(a * b, '\n')
print(np.multiply(a, b))
```

**Скалярное произведение векторов, умножение матриц:**

```python
a = np.array([2, 3])
b = np.array([[4], [5]])
```

```python
print(np.dot(a, b), '\n')
print(a @ b)
```

```python
print(np.dot(a, 3), '\n')
print(a * 3)
```

**7. Используя циклы написать собственную реализацию умножения двух двумерных матриц, засечь время выполнения и сравнить его с функцией умножения матриц NumPy:**

```python
n = 300
A = np.random.rand(n, n)
B = np.random.rand(n, n)
```

```python
%%time
C = A @ B
```

```python
%%time
C = np.zeros((n, n))
for i in range(n):
    for j in range(n):
        for k in range(n):
            C[i, j] += A[i, k] * B[k, j]
```
